### *Описание данных*

- **country (Страна)**: 
  - Название страны, для которой представлены данные о счастье и благополучии.

- **region (Регион)**: 
  - Регион, к которому принадлежит страна (например, Европа, Азия, Африка и т.д.).

- **happiness_score (Рейтинг счастья)**: 
  - Оценка уровня счастья в стране, основанная на различных факторах. Значения варьируются от 0 до 10, где более высокие значения указывают на более высокий уровень счастья.

- **gdp_per_capita (ВВП на душу населения)**: 
  - Экономический показатель, отражающий рыночную стоимость всех произведенных в стране товаров и услуг на душу населения. Измеряется в долларах США.

- **social_support (Социальная поддержка)**: 
  - Показатель, отражающий уровень социальной поддержки, доступной гражданам страны. Более высокие значения указывают на большую доступность помощи и поддержки.

- **healthy_life_expectancy (Ожидаемая продолжительность здоровой жизни)**: 
  - Средняя ожидаемая продолжительность жизни без серьезных заболеваний или инвалидности. Значения варьируются от 0 до 1.

- **freedom_to_make_life_choices (Свобода выбора)**: 
  - Показатель свободы выбора в жизни граждан страны. Более высокие значения указывают на большую степень личной свободы и возможности принимать решения.

- **generosity (Щедрость)**: 
  - Показатель щедрости граждан страны, основанный на данных о благотворительности и пожертвованиях. Значения варьируются от 0 до 1.

- **perceptions_of_corruption (Восприятие коррупции)**: 
  - Показатель восприятия коррупции в стране. Более высокие значения указывают на более низкое восприятие коррупции и большее доверие к правительству и институтам.ии и большее доверие к правительству и институтам.

```python
# Поиск стран с нулевыми значениями в social_support
countries_with_zero_social_support = df[df['social_support'] == 0]
print("Страны с нулевым значением социальной поддержки:")
print(countries_with_zero_social_support[['country', 'happiness_score', 'gdp_per_capita']])

# Поиск стран с нулевыми значениями в generosity
countries_with_zero_generosity = df[df['generosity'] == 0]
print("Страны с нулевым значением щедрости:")
print(countries_with_zero_generosity[['country', 'happiness_score', 'gdp_per_capita']])
```

```python
# Заполнение нулевых значений средним значением
mean_social_support = df['social_support'].mean()
df['social_support'].replace(0, mean_social_support, inplace=True)

mean_generosity = df['generosity'].mean()
df['generosity'].replace(0, mean_generosity, inplace=True)
```

```python
print("Количество пропущенных значений после заполнения:")
print(df.isna().sum())

print("Количество уникальных значений после заполнения:")
print(df['social_support'].value_counts())
print(df['generosity'].value_counts())
```

### Плотность распределения данных

```python
# График плотности для happiness_score
# График плотности (оценка ядра) показывает, как распределены значения переменной (например, уровень счастья) по непрерывной шкале. 
#Он создаёт плавную кривую, которая отображает вероятность нахождения значений в определённом диапазоне. 
#Этот график помогает визуализировать, где сосредоточены данные и есть ли у них пики.
plt.figure(figsize=(10, 6))
sns.kdeplot(df['happiness_score'], fill=True)
plt.title('Плотность распределения уровня счастья')
plt.xlabel('Уровень счастья')
plt.ylabel('Плотность')
plt.show()
```

```python
# этот график позволяет сравнить распределения нескольких переменных на одном графике. 
#Например можно увидеть, как распределяются ВВП на душу населения, социальная поддержка и ожидаемая продолжительность жизни.
#Это полезно для выявления взаимосвязей между различными показателями и понимания их совместного поведения.
plt.figure(figsize=(10, 6))
sns.kdeplot(df['gdp_per_capita'], label='GDP per Capita', fill=True)
sns.kdeplot(df['social_support'], label='Social Support', fill=True)
sns.kdeplot(df['healthy_life_expectancy'], label='Healthy Life Expectancy', fill=True)
plt.title('Плотность распределения для нескольких переменных')
plt.xlabel('Значение')
plt.ylabel('Плотность')
plt.legend()
plt.show()
```

```python
#Ящичковый график (ящик с усами) показывает основные статистические характеристики распределения переменной, такие как медиана, квартили и выбросы.
#Он помогает быстро оценить разброс данных и выявить аномалии. 
#Например, можно увидеть уровень счастья в разных регионах и определить, в каких регионах есть 
#значительные выбросы или различия в уровне счастья.
plt.figure(figsize=(10, 6))
sns.boxplot(x='region', y='happiness_score', data=df)
plt.title('Распределение уровня счастья по регионам')
plt.xticks(rotation=45)
plt.show()
```

```python
#график violin сочетает в себе элементы «box plot» и «kde plot». 
#Он показывает не только основные статистические показатели (медиану, квартили),
#но и плотность распределения данных. Это позволяет лучше понять, как распределены значения внутри каждой категории (например, уровень счастья по регионам). 
#График violin особенно полезен для сравнения нескольких групп.
plt.figure(figsize=(10, 6))
sns.violinplot(x='region', y='happiness_score', data=df, inner='quartile', palette='muted')
plt.title('Уровень счастья по регионам', fontsize=16)
plt.xlabel('Регион', fontsize=12)
plt.ylabel('Уровень счастья', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Добавление сетки для лучшего восприятия
plt.show()
```
